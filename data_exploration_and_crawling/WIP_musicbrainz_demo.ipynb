```python
import musicbrainzngs as m
m.set_useragent("application", "0.01", "http://example.com")
```

```python
def get_info(artist_row):
  data = {}
  data["spotify_id"] = artist_row.id 
  data["spotify_name"] = artist_row["name"]
  try:
    print("searching for", artist_row["name"])
    first_match = m.search_artists(artist_row["name"])["artist-list"][0]
    print("got data")
    data["mb_name"] = first_match.get("mb_name")
    data["country"] = first_match.get("country")
    data["gender"] = first_match.get("gender")
    data["mb_id"] = first_match.get("id") # store musicbrainz ID as well
  except:
    print("Could not get data for", artist_row["name"])
  return data
```

```python
artist_data_musicbrainz = artists[:10].apply(get_info, axis=1, result_type="expand")
```
