```python
cleaned_lyric_data = raw_lyric_data.copy()
# _type is song for all rows
cleaned_lyric_data.drop(columns="_type", inplace=True)

# all tracks are NOT instrumental
cleaned_lyric_data.drop(columns="instrumental", inplace=True)

# artist_names can also be reconstructed using featured_artists
cleaned_lyric_data.drop(columns="artist_names", inplace=True)

# full_title constructable title, artist and featured_artist (except for differences in characters used for apostrophes etc.)
cleaned_lyric_data.drop(columns="full_title", inplace=True)

# lyrics_state is complete for all lyrics
cleaned_lyric_data.drop(columns="lyrics_state", inplace=True)

# redundant - NaN for all rows
cleaned_lyric_data.drop(columns="0", inplace=True)

# release date info is not really reliable, better remove
cleaned_lyric_data.drop(
    columns=[
        "release_date_for_display",
        "release_date_with_abbreviated_month_for_display",
        "release_date_components",
    ],
    inplace=True,
)

# only page views (nested in this column's object) are potentially relevant - but change constantly as well, so better remove
cleaned_lyric_data.drop(columns="stats", inplace=True)

# title_with_featured reconstructable from title and featured_artists
# (except for differences in characters used for apostrophes etc. - like with full_title)
cleaned_lyric_data.drop(columns="title_with_featured", inplace=True)

# pyong is apparently whenever a user shares a lyric page with their followers
# pyong count pretty useless as a metric (number of pyongs for "most pyonged song" on Genius unknown)
cleaned_lyric_data.drop(columns="pyongs_count", inplace=True)

# need to clean lyrics as well
# all lyrics start with name of track followed by "Lyrics" and
# end with unnecessary "Embed" (probably unwanted artifact of page crawl)
# additionally, the phrase "You might also like" was added in random places for almost all the lyrics in the data collection process (don't know exactly how that happened)
cleaned_lyrics = (
    cleaned_lyric_data.lyrics.str.replace("^.*Lyrics\s*", "", regex=True)
    .str.replace("Embed$", "", regex=True)
    .str.replace("You might also like", "")
)

cleaned_lyric_data.lyrics = cleaned_lyrics
```

```python
out_path = create_data_out_path("genius_data_cleaned.json")
cleaned_lyric_data.to_json(out_path, orient="records", lines=True)
print("stored cleaned Genius data in", out_path)
```
