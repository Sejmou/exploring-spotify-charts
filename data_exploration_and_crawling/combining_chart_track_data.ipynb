```python
import pandas as pd
from helpers import get_data_path, create_data_out_path
```

```python
tracks = pd.read_csv(get_data_path("tracks.csv"), index_col="id").drop(
    columns=["uri", "type"] # URI can be recreated from id, type is "track" for all tracks
)
tracks
```

```python
track_features = pd.read_csv(get_data_path("audio_features.csv"), index_col="id").drop(
    columns=["uri", "type"] # URI can be recreated from id, type is "track" for all tracks
)
track_features
```

```python
track_and_artist = (
    pd.read_csv(get_data_path("track_and_primary_artist_names.csv"))
    .rename(
        columns={
            "track_id": "id",
            "artist_id": "primary_artist_id",
            "artist": "primary_artist_name",
        }
    )
    .set_index("id")
)
track_and_artist
```

```python
track_data = tracks.join(track_and_artist).join(track_features, lsuffix="_base", rsuffix="_features")
track_data
```

```python
different_durations = track_data.loc[
    track_data.duration_ms_base != track_data.duration_ms_features,
    ["duration_ms_base", "duration_ms_features"],
]
different_durations
```

```python
differences = (
    different_durations.duration_ms_base - different_durations.duration_ms_features
)
differences
```

```python
differences.loc[differences == -1.0]
```

```python
differences.loc[differences != -1.0]
```

Hmm, the durations differ by exactly 1 for all data except a very small subset. Let's just stick with the ones from the track_features dataframe.


```python
track_data = track_data.drop(columns="duration_ms_base").rename(
    columns={"duration_ms_features": "duration_ms"}
)
track_data
```
